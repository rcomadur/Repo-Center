```python
import pandas as pd 
import matplotlib.pyplot as plt 

# run plots in the notebook 
%matplotlib inline 
url = "http://pbpython.com/extras/sample-salesv2.csv" 
sales = pd.read_csv(url)
```

```python
print (sales.head())
```

```python
sales.columns = ['account_number', 'name', 'sku', 'category','quantity','unit_price','ext_price', 'date']
```

```python
print (sales.head())
```

```python
print (sales[['name', 'category','quantity','unit_price']][:])
```

```python
#Subset the dataframe to contain only shirt sales
```

```python
tshirst = sales[sales['category'] == "Shirt"]
```

```python
print(tshirst)
```

```python
tshirt_sub=  tshirst[['name', 'category','quantity','unit_price']][:]
```

```python
print(tshirt_sub)
```

```python
tshirt_sub['cost per shirt'] = tshirt_sub['unit_price'] / tshirt_sub['quantity']
```

```python
print(tshirt_sub)
```

```python
print (tshirt_sub)
```

```python
tshirt_sub2 = (tshirt_sub.groupby('name').mean())
```

```python
print (tshirt_sub2)
```

```python
tshirtsub_3= tshirt_sub2.sort_values(by=['cost per shirt'], ascending=False)
```

```python
print(tshirtsub_3[1:10])
```
